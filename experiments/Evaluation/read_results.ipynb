```python
import pandas as pd

df = pd.read_json("results4.json")
df_stats_original = df['flopco_stats_original'].apply(lambda x: pd.Series(x))
df_stats_compressed = df['flopco_stats_compressed'].apply(lambda x: pd.Series(x))
df_stats = pd.concat([df_stats_original, df_stats_compressed], axis=1, keys=['original', 'compressed'])
df = pd.concat([df, df_stats], axis=1)
df.drop(columns=['flopco_stats_original', 'flopco_stats_compressed'], inplace=True)
df_compression_ratio = df['compression_ratio'].apply(lambda x: pd.Series(x))
df = pd.concat([df, df_compression_ratio], axis=1)
df.drop(columns=['compression_ratio'], inplace=True)
```
