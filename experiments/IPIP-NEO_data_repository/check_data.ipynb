```python
import pandas as pd
```

```python
# Указываем схему ширины колонок (colspecs)
colspecs = [
    (0, 6),  # CASE
    (6, 7),  # SEX
    (7, 9),  # AGE
    (9, 11),  # SEC
    (11, 13),  # MIN
    (13, 15),  # HOUR
    (15, 17),  # DAY
    (17, 19),  # MONTH
    (19, 22),  # YEAR
    (22, 31),  # COUNTRY
]

# Добавляем колонки для I1-I120
for i in range(120):
    colspecs.append((31 + i * 1, 31 + (i + 1) * 1))  # noqa: PERF401
    # Каждое поле I1-I120 занимает 1 символ
```

```python
data_path = "./osfstorage-archive (2)/IPIP120.dat"
```

```python
# Чтение данных с фиксированными ширинами
data = pd.read_fwf(data_path, colspecs=colspecs, header=None)
```

```python
# Присваиваем имена колонок
columns = ["CASE", "SEX", "AGE", "SEC", "MIN", "HOUR", "DAY", "MONTH", "YEAR", "COUNTRY"]
columns += [f"I{i+1}" for i in range(120)]  # Имена для I1 - I120
data.columns = columns
```

```python
int_columns = ["CASE", "SEX", "AGE", "SEC", "MIN", "HOUR", "DAY", "MONTH", "YEAR"]
int_columns += [f"I{i+1}" for i in range(120)]

for col in int_columns:
    data[col] = pd.to_numeric(data[col]).astype(int)
```
